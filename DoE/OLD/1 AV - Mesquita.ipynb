## Análise de Variância

### Análise Geral

```r
tab <- read.csv(file = './DoE/DoE - Results v5.csv', sep = ",", dec = ".")
head(tab)
```

```r
TIC<-tab$Cost
alg<-factor(tab$algorithm)
n<-factor(tab$N)
cv<-factor(tab$cv)
rho<-factor(tab$rho)
p<-factor(tab$p)
```

```r
av<-aov(TIC~n*cv*rho*p*alg)
summary(av)
```

```r
par(bg = "#f7f7f7")
boxplot(TIC)
```

```r
par(bg = "#f7f7f7")
boxplot(TIC~alg)
```

```r
par(bg = "#f7f7f7")
boxplot(TIC~n*alg)
```
